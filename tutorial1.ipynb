```python
from Bio.Seq import Seq

my_seq = Seq("GATCG")
for i, l in enumerate(my_seq):
    print("%i %s" % (i, l))
```

```python
my_seq = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC")
len(my_seq)
```

```python
g_count = my_seq.count("G")
c_count = my_seq.count("C")


percentage = (g_count + c_count)/len(my_seq) * 100
print(percentage)
```
