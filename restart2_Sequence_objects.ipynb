## 1) `Seq` objects act like strings
A `Seq` object stores biological sequence letters and supports many string-like operations.
You can get length, index letters, and loop through letters.

##### Seq object is in some ways “read-only”. If you need to edit your sequence, for example simulating a point mutation, look at the Section MutableSeq objects below which talks about the MutableSeq object.

```python
# Import the Seq class from Biopython
from Bio.Seq import Seq
# Create a DNA sequence object
my_seq=Seq("GATCG")
# Print the full sequence
print("Sequence:", my_seq)
# len(...) returns number of letters
print("Length:", len(my_seq))

# Indexing works like Python strings (0-based indexing)
print("First letter (index 0):", my_seq[0])
print("Third letter (index 1):", my_seq[1])
print("Third letter (index 2):", my_seq[2])
print("Third letter (index 3):", my_seq[3])
print("Third letter (index 4:", my_seq[4])

print("Last letter (index -2):", my_seq[-2])

```

```python
# enumerate(...) gives index + value while looping
for index, letter in enumerate(my_seq):
    print("Index", index, "has", letter)
```

## 2) Counting letters and GC fraction
Use `.count()` to count letters or motifs. For GC content, use `gc_fraction` from `Bio.SeqUtils`.

```python
from Bio.Seq import Seq # 
from Bio.SeqUtils import gc_fraction #Calculate G+C percentage in seq (float between 0 and 1).

# Example sequence
seq_name = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC") 
#this means that we will count the number of AA's that appeared together in the sequence, but not overlapping (e.g. AAAA has 2 AA's, not 3), when  it is AAA it will count 1 AA, not 2 and ignore the A's that is not in pairs or are not together

print("Length:",len(seq_name))
# Count single-letter motif
print("Count of G:", seq_name.count("G"))

# Count short motif (non-overlapping behavior)
print("Count of AA (non-overlapping):", seq_name.count("AA"))


# Calculate GC fraction (0 to 1)
#gc means that we will count the number of G's and C's in the sequence and divide it by the total length of the sequence to get a fraction that represents how much of the sequence is made up of G's and C's, which is important because G-C pairs are stronger than A-T pairs in DNA, so a higher GC content can indicate a more stable DNA molecule
print("Count of G:", seq_name.count("G"))
print("Count of C:", seq_name.count("C"))

print("Count of GC:", seq_name.count("GC"))

gc = gc_fraction(seq_name)
print("GC fraction:", gc) # 9 + 6 =15, so 15/32 = 0.46875, that is how come we got out gc_fraction
print("GC percent:", gc * 100)
```

## 3) Slicing, stride, and reversing
Slicing uses `[start:stop:step]`. Start is included; stop is excluded.
A negative step (like `-1`) reverses the sequence.

```python
from Bio.Seq import Seq

seq = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC")
# Basic slice: characters from index 4 up to (not including) 12
print("seq[4:12] =>", seq[4:12])
# Every 3rd base from different starts (codon positions)
print("Position 1 bases:", seq[0::3]) # this means that we will start from index 0 and take every 3rd base, so we will get the bases at index 0, 3, 6, 9, etc. which are the first bases of each codon in a DNA sequence. this will move to the  +3 to get the next base in the same position of the next codon, so we will get the first bases of each codon in the sequence, this is useful for analyzing codon usage and understanding how the genetic code is read in groups of three bases (codons) to specify amino acids during protein synthesis.

print("Position 2 bases:", seq[1::3]) # this means that we will start from index 1 and take every 3rd base, so we will get the bases at index 1, 4, 7, 10, etc. which are the second bases of each codon in a DNA sequence, this means

print("Position 3 bases:", seq[2::3])# this means that we will start from index 2 and take every 3rd base, so we will get the bases at index 2, 5, 8, 11, etc. which are the third bases of each codon in a DNA sequence

# Reverse sequence with step -1
print("Reversed:", seq[::-1])
```

## 4) Strings, concatenation, join, and case
You can convert `Seq` to plain strings with `str(...)`, add sequences with `+`, and use `.join(...)` with spacers.

```python
from Bio.Seq import Seq

a = Seq("ACGT")
b = Seq("AACCGG")

# Convert to normal Python string
print("As string:", str(a))

# Concatenate two sequences
print("a + b =>", a + b)

# Join several contigs using an N spacer
contigs = [Seq("ATG"), Seq("ATCCCG"), Seq("TTGCA")]
spacer = Seq("N" * 5)
print("Joined contigs:", spacer.join(contigs))

# Case conversion
mixed = Seq("acgtACGT")
print("Upper:", mixed.upper())
print("Lower:", mixed.lower())
```

## 5) Complement, reverse complement, transcription, and back-transcription
For DNA: complement swaps A↔T and C↔G. Reverse complement reverses then complements.
Transcription converts DNA `T` to RNA `U`. Back-transcription converts `U` back to `T`.

```python
from Bio.Seq import Seq

# Example coding DNA strand
coding_dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
# transcription is the process of converting DNA to RNA, where the DNA sequence is used as a template to create a complementary RNA sequence, in this process, the base thymine (T) in DNA is replaced by uracil (U) in RNA, so when we transcribe the coding_dna sequence, we will get an mRNA sequence where all the T's are replaced by U's, this is important because mRNA is what is used as a template for protein synthesis during translation, and it allows the genetic information stored in DNA to be expressed as proteins that perform various functions in the cell.

# translation is the process of converting mRNA into a protein sequence, where the mRNA sequence is read in groups of three bases (codons) and each codon specifies a particular amino acid, so when we translate the coding_dna sequence, we will first transcribe it into mRNA and then read the mRNA sequence in codons to determine the corresponding amino acids that will be linked together to form a protein, this is how the genetic information stored in DNA is ultimately expressed as functional proteins in the cell.
# Complement and reverse complement
print("Complement:", coding_dna.complement())
print("Reverse complement:", coding_dna.reverse_complement())

# DNA -> RNA (T becomes U)
mrna = coding_dna.transcribe()
print("mRNA:", mrna)

# RNA -> DNA (U becomes T)
print("Back-transcribed DNA:", mrna.back_transcribe())
```

## 6) Translation and genetic code tables
Translation converts nucleotide triplets (codons) into amino acids.
Different organisms can use different genetic code tables.

```python
from Bio.Seq import Seq
from Bio.Data import CodonTable

dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")

# Default translation (table 1: Standard)
print("Default translate:", dna.translate())

# Stop at first in-frame stop codon
print("to_stop=True:", dna.translate(to_stop=True))

# Use vertebrate mitochondrial table (NCBI id 2)
print("Table 2 translate:", dna.translate(table=2))
print("Table 2 to_stop=True:", dna.translate(table=2, to_stop=True))

# Show table metadata
mito = CodonTable.unambiguous_dna_by_id[2]
print("Mito stop codons:", mito.stop_codons)
print("Mito start codons:", mito.start_codons)
print("ACG maps to amino acid:", mito.forward_table["ACG"])

fasta_format_string = ">Name\n%s\n" % dna
print(fasta_format_string)
```

## 7) Comparing sequences, unknown sequences, and partially defined sequences
`Seq` comparisons are based on sequence letters.
Biopython also supports undefined (`Seq(None, length)`) and partially defined sequences.

```python
from Bio.Seq import Seq

# Comparison with string and Seq
seq1 = Seq("ACGT")
print("Seq == string:", seq1 == "ACGT")
print("string == Seq:", "ACGT" == seq1)

# Unknown sequence: length known, content undefined
unknown = Seq(None, 10)
print("Unknown object:", unknown)
print("Unknown length:", len(unknown))

# Partially defined sequence: one known region in a long sequence
partial = Seq({117512683: "TTGAAAACCTGAATGTGAGAGTCAGTCAAGGATAGT"}, length=159345973)
print("Defined slice:", partial[117512690:117512700])
print("Undefined slice:", partial[1000:1020])
```

## 8) MutableSeq, finding subsequences, and function-style utilities
`Seq` is immutable (read-only). Use `MutableSeq` when you want to edit sequence letters.
You can find motifs with `.find()`, `.rfind()`, `.index()`, and `.search()`.

```python
from Bio.Seq import Seq, MutableSeq, reverse_complement, transcribe, back_transcribe, translate

# MutableSeq lets you edit letters
m = MutableSeq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
print("Original MutableSeq:", m)

# Change one letter at index 5
m[5] = "C"
print("After substitution:", m)

# Remove first occurrence of T
m.remove("T")
print("After removing one T:", m)

# Reverse in place
m.reverse()
print("Reversed MutableSeq:", m)

# Convert back to immutable Seq
read_only_seq = Seq(m)
print("Immutable Seq:", read_only_seq)

# Subsequence searching
seq = Seq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
print("find('CC'):", seq.find("CC"))
print("rfind('CC'):", seq.rfind("CC"))
print("find('ACTG') when missing:", seq.find("ACTG"))

for index, sub in seq.search(["CC", "GGG", "CC"]):
    print("search found:", index, sub)

# Function-style equivalents (work on plain strings too)
dna_string = "GCTGTTATGGGTCGTTGGAAGGGTGGTCGTGCTGCTGGTTAG"
print("reverse_complement:", reverse_complement(dna_string))
print("transcribe:", transcribe(dna_string))
print("back_transcribe:", back_transcribe(transcribe(dna_string)))
print("translate:", translate(dna_string))
```
