## 1) `Seq` objects act like strings
A `Seq` object stores biological sequence letters and supports many string-like operations.
You can get length, index letters, and loop through letters.

##### Seq object is in some ways “read-only”. If you need to edit your sequence, for example simulating a point mutation, look at the Section MutableSeq objects below which talks about the MutableSeq object.

```python
# Import the Seq class from Biopython
from Bio.Seq import Seq
# Create a DNA sequence object
my_seq=Seq("GATCG")
# Print the full sequence
print("Sequence:", my_seq)
# len(...) returns number of letters
print("Length:", len(my_seq))

# Indexing works like Python strings (0-based indexing)
print("First letter (index 0):", my_seq[0])
print("Third letter (index 1):", my_seq[1])
print("Third letter (index 2):", my_seq[2])
print("Third letter (index 3):", my_seq[3])
print("Third letter (index 4:", my_seq[4])

print("Last letter (index -2):", my_seq[-2])

```

```python
# enumerate(...) gives index + value while looping
for index, letter in enumerate(my_seq):
    print("Index", index, "has", letter)
```

## 2) Counting letters and GC fraction
Use `.count()` to count letters or motifs. For GC content, use `gc_fraction` from `Bio.SeqUtils`.

```python
from Bio.Seq import Seq # 
from Bio.SeqUtils import gc_fraction #Calculate G+C percentage in seq (float between 0 and 1).

# Example sequence
seq_name = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC") 
#this means that we will count the number of AA's that appeared together in the sequence, but not overlapping (e.g. AAAA has 2 AA's, not 3), when  it is AAA it will count 1 AA, not 2 and ignore the A's that is not in pairs or are not together

print("Length:",len(seq_name))
# Count single-letter motif
print("Count of G:", seq_name.count("G"))

# Count short motif (non-overlapping behavior)
print("Count of AA (non-overlapping):", seq_name.count("AA"))


# Calculate GC fraction (0 to 1)
#gc means that we will count the number of G's and C's in the sequence and divide it by the total length of the sequence to get a fraction that represents how much of the sequence is made up of G's and C's, which is important because G-C pairs are stronger than A-T pairs in DNA, so a higher GC content can indicate a more stable DNA molecule
print("Count of G:", seq_name.count("G"))
print("Count of C:", seq_name.count("C"))

print("Count of GC:", seq_name.count("GC"))

gc = gc_fraction(seq_name)
print("GC fraction:", gc) # 9 + 6 =15, so 15/32 = 0.46875, that is how come we got out gc_fraction
print("GC percent:", gc * 100)
```

## 3) Slicing, stride, and reversing
Slicing uses `[start:stop:step]`. Start is included; stop is excluded.
A negative step (like `-1`) reverses the sequence.

```python
from Bio.Seq import Seq

seq = Seq("GATCGATGGGCCTATATAGGATCGAAAATCGC")
# Basic slice: characters from index 4 up to (not including) 12
print("seq[4:12] =>", seq[4:12])
# Every 3rd base from different starts (codon positions)
print("Position 1 bases:", seq[0::3]) # this means that we will start from index 0 and take every 3rd base, so we will get the bases at index 0, 3, 6, 9, etc. which are the first bases of each codon in a DNA sequence. this will move to the  +3 to get the next base in the same position of the next codon, so we will get the first bases of each codon in the sequence, this is useful for analyzing codon usage and understanding how the genetic code is read in groups of three bases (codons) to specify amino acids during protein synthesis.

print("Position 2 bases:", seq[1::3]) # this means that we will start from index 1 and take every 3rd base, so we will get the bases at index 1, 4, 7, 10, etc. which are the second bases of each codon in a DNA sequence, this means

print("Position 3 bases:", seq[2::3])# this means that we will start from index 2 and take every 3rd base, so we will get the bases at index 2, 5, 8, 11, etc. which are the third bases of each codon in a DNA sequence

# Reverse sequence with step -1
print("Reversed:", seq[::-1])
```

## 4) Strings, concatenation, join, and case
You can convert `Seq` to plain strings with `str(...)`, add sequences with `+`, and use `.join(...)` with spacers.

```python
from Bio.Seq import Seq

a = Seq("ACGT")
b = Seq("AACCGG")

# Convert to normal Python string
print("As string:", str(a))

# Concatenate two sequences
print("a + b =>", a + b)

# Join several contigs using an N spacer
contigs = [Seq("ATG"), Seq("ATCCCG"), Seq("TTGCA")]
spacer = Seq("N" * 5)
print("Joined contigs:", spacer.join(contigs))

# Case conversion
mixed = Seq("acgtACGT")
print("Upper:", mixed.upper())
print("Lower:", mixed.lower())
```

## 5) Complement, reverse complement, transcription, and back-transcription
For DNA: complement swaps A↔T and C↔G. Reverse complement reverses then complements.
Transcription converts DNA `T` to RNA `U`. Back-transcription converts `U` back to `T`.

```python
from Bio.Seq import Seq

# Example coding DNA strand
coding_dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
# transcription is the process of converting DNA to RNA, where the DNA sequence is used as a template to create a complementary RNA sequence, in this process, the base thymine (T) in DNA is replaced by uracil (U) in RNA, so when we transcribe the coding_dna sequence, we will get an mRNA sequence where all the T's are replaced by U's, this is important because mRNA is what is used as a template for protein synthesis during translation, and it allows the genetic information stored in DNA to be expressed as proteins that perform various functions in the cell.

# translation is the process of converting mRNA into a protein sequence, where the mRNA sequence is read in groups of three bases (codons) and each codon specifies a particular amino acid, so when we translate the coding_dna sequence, we will first transcribe it into mRNA and then read the mRNA sequence in codons to determine the corresponding amino acids that will be linked together to form a protein, this is how the genetic information stored in DNA is ultimately expressed as functional proteins in the cell.
# Complement and reverse complement
print("Complement:", coding_dna.complement())
print("Reverse complement:", coding_dna.reverse_complement())

# DNA -> RNA (T becomes U)
mrna = coding_dna.transcribe()
print("mRNA:", mrna)

# RNA -> DNA (U becomes T)
print("Back-transcribed DNA:", mrna.back_transcribe())
```

```python
from Bio.Data import CodonTable
from Bio.Seq import Seq

def analyze_translation_table(table_id):
    """Analyze and explain a translation table"""
    try:
        table = CodonTable.unambiguous_dna_by_id[table_id]
        
        print(f"\n{'='*60}")
        print(f"TABLE {table_id}: {table.names[0]}")
        print(f"{'='*60}")
        
        # Alternative names
        if len(table.names) > 1:
            print(f"Also known as: {', '.join(table.names[1:])}")
        
        # Key characteristics
        print(f"\n📌 KEY FEATURES:")
        print(f"  • Start codons: {', '.join(table.start_codons[:5])}" + 
              (f"..." if len(table.start_codons) > 5 else ""))
        print(f"  • Stop codons: {', '.join(table.stop_codons)}")
        
        # Find what's DIFFERENT from standard code (table 1)
        if table_id != 1:
            standard = CodonTable.unambiguous_dna_by_id[1]
            differences = []
            
            # Check all codons
            for codon in table.forward_table.keys():
                if codon in standard.forward_table:
                    if table.forward_table[codon] != standard.forward_table[codon]:
                        differences.append(f"{codon}: {standard.forward_table[codon]}→{table.forward_table[codon]}")
            
            # Check stop codons
            for codon in table.stop_codons:
                if codon in standard.forward_table:
                    differences.append(f"{codon}: {standard.forward_table[codon]}→STOP")
                elif codon not in standard.stop_codons:
                    differences.append(f"{codon}: (new stop codon)")
            
            if differences:
                print(f"\n🔄 DIFFERENCES FROM STANDARD CODE:")
                for diff in differences[:8]:  # Show first 8
                    print(f"  • {diff}")
                if len(differences) > 8:
                    print(f"  • ... and {len(differences)-8} more")
        
        # Show a practical example
        print(f"\n🧪 PRACTICAL EXAMPLE:")
        # Create a test sequence that highlights differences
        test_seq = create_test_sequence(table_id)
        dna = Seq(test_seq)
        print(f"  DNA: {dna}")
        print(f"  Protein: {dna.translate(table=table_id)}")
        
        return table
    except:
        print(f"Table {table_id} not available")
        return None

def create_test_sequence(table_id):
    """Create a test sequence that shows unique features of each table"""
    
    # Base sequence with common codons
    base = "ATGGCCATTGTA"  # M A I V
    
    # Table-specific problematic codons
    problematic = {
        1: "TGATAA",      # Standard stops
        2: "TGATAA",      # UGA=W in mito
        3: "TGATAA",      # Yeast mito
        4: "TGATAA",      # Mold mito
        5: "TGATAA",      # Invert mito
        6: "TAA",         # Ciliate - TAA=Q
        9: "TGA",         # Alternative flatworm
        10: "TGATAA",     # Euplotid
        11: "TGATAA",     # Bacterial
        12: "TGA",        # Alternative yeast
        13: "TGA",        # Ascidian mito
        14: "TGA",        # Flatworm mito
        15: "TGA",        # Blepharisma
        16: "TGA",        # Chlorophycean mito
        21: "TGA",        # Trematode mito
        22: "TGA",        # Scenedesmus mito
        23: "TGA",        # Thraustochytrium mito
        24: "TGA",        # Pterobranchia mito
        25: "TGA",        # Candidate Division SR1
    }
    
    extra = problematic.get(table_id, "TGATAA")
    return base + extra + "GGGCCC"

# Analyze ALL tables (1-31)
print("🧬 COMPLETE GENETIC CODE TABLES REFERENCE")
print("=" * 60)

for i in range(1, 32):
    analyze_translation_table(i)
```

## 6) Translation and genetic code tables
Translation converts nucleotide triplets (codons) into amino acids.
Different organisms can use different genetic code tables.

```python
from Bio.Seq import Seq
from Bio.Data import CodonTable

dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")

# Default translation (table 1: Standard)
print("Default translate:", dna.translate()) #meaning that we will take the DNA sequence and read it in groups of three bases (codons) to determine the corresponding amino acids that will be linked together to form a protein, this is how the genetic information stored in DNA is ultimately expressed as functional proteins in the cell, the default translation uses the standard genetic code (table 1) which is the most common code used by most organisms, but there are other genetic codes (e.g. mitochondrial codes) that can be specified using the table parameter in the translate method.

# Stop at first in-frame stop codon
print("to_stop=True:", dna.translate(to_stop=True))

# Use vertebrate mitochondrial table (NCBI id 2)
print("Table 2 translate:", dna.translate(table=2))
print("Table 2 to_stop=True:", dna.translate(table=2, to_stop=True))

# Show table metadata
mito = CodonTable.unambiguous_dna_by_id[2]
print("Mito stop codons:", mito.stop_codons)
print("Mito start codons:", mito.start_codons)
print("ACG maps to amino acid:", mito.forward_table["ACG"])

fasta_format_string = ">Name\n%s\n" % dna
print(fasta_format_string)


coding_dna = Seq("UCAGCACGGCAG")
mrna = coding_dna.translate()
print("mRNA:", mrna)
```

## 7) Comparing sequences, unknown sequences, and partially defined sequences
`Seq` comparisons are based on sequence letters.
Biopython also supports undefined (`Seq(None, length)`) and partially defined sequences.

```python
from Bio.Seq import Seq

# Comparison with string and Seq
seq1 = Seq("ACGT")
print("Seq == string:", seq1 == "ACGT")
print("string == Seq:", "ACGT" == seq1)

# Unknown sequence: length known, content undefined
unknown = Seq(None, 10) # this means that we will create a Seq object called unknown where the content of the sequence is not defined (None) but we know that it has a length of 10, this can be useful for representing sequences where we only know the length but not the actual sequence, such as when we have incomplete data or when we want to represent a placeholder for a sequence that will be determined later. The unknown sequence will have a length of 10, but its content will be undefined until we assign a specific sequence to it.
print("Unknown object:", repr(unknown))
print("Unknown length:", len(unknown))

# Partially defined sequence: one known region in a long sequence
partial = Seq({117512683: "TTGAAAACCTGAATGTGAGAGTCAGTCAAGGATAGT"}, length=159345973)# this means that we will create a Seq object called partial where we have one known region of the sequence defined at a specific index (117512683) with the sequence "TTGAAAACCTGAATGTGAGAGTCAGTCAAGGATAGT", and the total length of the sequence is 159345973, this allows us to represent a sequence where we only have information about a specific region while the rest of the sequence is undefined, we can access the defined region using slicing, and any slice that includes undefined regions will return None or an appropriate representation of undefined content.
print("Defined slice:", partial[117512690:117512700])
print("Undefined slice:", repr(partial[1000:1020]))
```

## 8) MutableSeq, finding subsequences, and function-style utilities
`Seq` is immutable (read-only). Use `MutableSeq` when you want to edit sequence letters.
You can find motifs with `.find()`, `.rfind()`, `.index()`, and `.search()`.

```python
from Bio.Seq import Seq, MutableSeq, reverse_complement, transcribe, back_transcribe, translate # Importing the Seq and MutableSeq classes, as well as some utility functions for working with sequences, this means that we will be able to use the Seq class to create immutable sequence objects that cannot be modified after they are created, and the MutableSeq class to create mutable sequence objects that can be modified after they are created, we will also have access to utility functions like reverse_complement, transcribe, back_transcribe, and translate that allow us to perform common operations on sequences such as getting the reverse complement of a sequence, transcribing DNA to RNA, back-transcribing RNA to DNA, and translating DNA or RNA to a protein sequence, these functions are essential for working with biological sequences and analyzing their properties and functions in the context of molecular biology and bioinformatics.

# MutableSeq lets you edit letters
m = MutableSeq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
print("Original MutableSeq:", m) # this means that we have created a mutable sequence object called m with the sequence "GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA", and we can modify this sequence by changing individual letters, adding or removing letters, or performing other operations that would not be possible with an immutable Seq object, this is useful for situations where we need to make changes to a sequence after it has been created, such as when we are simulating mutations or editing sequences in a laboratory setting.

# Change one letter at index 5
m[5] = "C"
print("After substitution:", m) # this means that we will change the letter at index 5 of the mutable sequence m to "C", so if the original letter at index 5 was something else (e.g. "T"), it will now be replaced with "C", this is useful for simulating point mutations or editing specific bases in a sequence to see how it affects the overall sequence and its properties.

# Remove first occurrence of T
m.remove("T")
print("After removing one T:", m) # this means that we will remove the first occurrence of the letter "T" from the mutable sequence m, so if there are multiple "T"s in the sequence, only the first one will be removed, this is useful for simulating mutations or editing sequences where we want to remove a specific base from the sequence. Ater removing the first "T", the sequence will be updated to reflect this change, and we can see how the sequence looks after the modification. and it  the original m="GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA" will become             "GCCACCGTAATGGGCCGCTGAAAGGGTGCCCGA" after the substitution and then "GCCACCGAAATGGGCCGCTGAAAGGGTGCCCGA" after removing the first "T".

# Reverse in place
m.reverse()
print("Reversed MutableSeq:", m) # this means that we will reverse the order of the letters in the mutable sequence m in place, so the first letter will become the last letter, the second letter will become the second to last letter, and so on, this is useful for analyzing sequences in reverse orientation or simulating processes like reverse transcription where the sequence is read in reverse.

# Convert back to immutable Seq
read_only_seq = Seq(m)
print("Immutable Seq:", read_only_seq) # this means that we will take the mutable sequence m and convert it back to an immutable Seq object called read_only_seq, which means that we will no longer be able to modify the sequence after this point, this is useful for situations where we want to ensure that a sequence cannot be changed after it has been created, such as when we are working with reference sequences or when we want to prevent accidental modifications to a sequence that is being used in multiple places in our code. to understand this, we can think of the mutable sequence m as a draft version of a sequence that we can edit and modify as needed, and once we are satisfied with the changes we have made, we can convert it to an immutable Seq object (read_only_seq) to ensure that it cannot be changed anymore, this is a common workflow in bioinformatics where we may need to make several edits to a sequence before finalizing it for analysis or use in downstream applications.

# Subsequence searching
seq = Seq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
print("find('CC'):", seq.find("CC")) # this means that we will search for the first occurrence of the subsequence "CC" in the sequence seq, and return the index of the first letter of that occurrence, if "CC" is found in the sequence, it will return the index where it starts, if it is not found, it will return -1, this is useful for finding specific motifs or patterns in a sequence, such as restriction enzyme recognition sites or other functional elements.

print("rfind('CC'):", seq.rfind("CC")) # this means that we will search for the last occurrence of the subsequence "CC" in the sequence seq, and return the index of the first letter of that occurrence, if "CC" is found in the sequence, it will return the index where it starts, if it is not found, it will return -1, this is useful for finding specific motifs or patterns in a sequence, such as restriction enzyme recognition sites or other functional elements, but starting from the end of the sequence instead of the beginning.
print("find('ACTG') when missing:", seq.find("ACTG"))

for index, sub in seq.search(["CC", "GGG", "CC"]):
    print("search found:", index, sub)

# Function-style equivalents (work on plain strings too)
dna_string = "GCTGTTATGGGTCGTTGGAAGGGTGGTCGTGCTGCTGGTTAG"
print("reverse_complement:", reverse_complement(dna_string)) # this means that we will take the dna_string and compute its reverse complement, which involves first finding the complement of each base (A <-> T, C <-> G) and then reversing the order of the bases in the resulting sequence, this is useful for analyzing sequences in reverse orientation or simulating processes like reverse transcription where the sequence is read in reverse. the output will be the reverse complement of the original dna_string, which can be used for various analyses in molecular biology and bioinformatics.
print("transcribe:", transcribe(dna_string))
print("back_transcribe:", back_transcribe(transcribe(dna_string)))
print("translate:", translate(dna_string))
```

### Basic Subsequence Finding

```python
from Bio.Seq import Seq

# Create a DNA sequence
dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
print(f"Sequence:", dna)
print(f"Length: {len(dna)}")

# Find a simple subsequence
pos = dna.find("AAAGGGTGCC")
print(f"'AAAGGGTGCC' found at position: {pos}")

# What if it's not there?
pos = dna.find("XYZ")  # Doesn't exist
print(f"'XYZ' found at position: {pos}")  # Returns -1

# Find starting from a specific position
pos = dna.find("GG", 10)  # Start searching from position 10
print(f"'GG' found after position 10 at: {pos}")

# Find the last occurrence
pos = dna.rfind("GG")
print(f"Last 'GG' found at: {pos}")
```
