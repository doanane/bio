# seqRecord is a SeqRecord object, which is a container for a sequence and its associated information. It is part of the Biopython library and is used to store and manipulate biological sequences along with their metadata. A SeqRecord object typically contains a Seq object (the actual sequence), an identifier (id), a name, a description, and any additional annotations or features related to the sequence. This allows for more comprehensive handling of biological data, as it can store not only the sequence itself but also relevant information about it.

```python
from Bio.Seq import Seq 
# Importing the Seq class from the Bio.Seq module to work with biological sequences.
from Bio.SeqRecord import SeqRecord #importing the SeqRecord class from the Bio.SeqRecord module to create and manipulate sequence records.
from Bio.SeqFeature import SeqFeature, FeatureLocation # Importing the SeqFeature and FeatureLocation classes from the Bio.SeqFeature module to define and manipulate features within a sequence record.
from Bio import SeqIO # Importing the SeqIO module from Biopython to read and write sequence records in various file formats.

# 1) Create a SeqRecord
 #records are used to store a sequence along with its associated metadata, such as an identifier, name, and description.
#  The SeqRecord class provides a convenient way to organize and manage sequence data, allowing you to easily access and manipulate both the sequence and its metadata.
record = SeqRecord(
    Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG"), # Creating a SeqRecord object with a DNA sequence.
    id="SEQ001",
    name="ExampleGene",
    description="Demo sequence record"
)
```

```python
print("ID:", record.id)
print("Name:", record.name)
print("Description:", record.description)
```

## Connecting with biological databases (NCBI Entrez)
Biopython provides `Entrez` to query NCBI databases (nucleotide, protein, pubmed, etc.).
This workflow has 3 common steps:
1. **Search** for record IDs with `Entrez.esearch`
2. **Fetch** data with `Entrez.efetch`
3. **Parse** the result (as text, XML, or sequence records)

> NCBI requires an email address so they can contact you if needed. Use your real email before running queries.

```python
# Step-by-step NCBI access using Biopython Entrez
from Bio import Entrez, SeqIO # Importing the Entrez and SeqIO modules from Biopython. The Entrez module is used to access the NCBI databases, while SeqIO is used for reading and writing sequence data.

from urllib.error import HTTPError # Importing the HTTPError class from the urllib.error module to handle HTTP errors that may occur when accessing the NCBI databases using the Entrez module. This allows us to catch and handle any errors that may arise during the communication with the NCBI servers, such as network issues or invalid requests.

# 1) Tell NCBI who you are (required)
Entrez.email = "anane365221@gmail.com.com"

search_term = "Homo sapiens[Organism] AND BRCA1[Gene]" 
# search  the term homo sapien

try:
    # 2) Search in nucleotide database
    with Entrez.esearch(db="nucleotide", term=search_term, retmax=5) as handle: 
        search_results = Entrez.read(handle)

    id_list = search_results["IdList"]
    print("Found IDs:", id_list)

    if not id_list:
        print("No results found for this query.")
    else:
        # 3) Fetch the first result in FASTA format
        first_id = id_list[0]
        print("\nFetching first ID:", first_id)

        with Entrez.efetch(db="nucleotide", id=first_id, rettype="fasta", retmode="text") as handle:
            seq_record = SeqIO.read(handle, "fasta")

        print("Sequence ID:", seq_record.id)
        print("Length:", len(seq_record))
        print("Description:", seq_record.description)
        print("First 120 bases:\n", seq_record.seq[:120])

except HTTPError as err:
    print("HTTP error from NCBI:", err)
except Exception as err:
    print("Unexpected error:", err)
```
