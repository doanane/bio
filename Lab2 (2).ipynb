## Part 1: Basic Global Alignment

Use the following sequences:

- seq1 = "AAAGGTCCTAGTTGG"
- seq2 = "AAAGTCCTAGTTGG"

```python
# TODO:
# 1. Align seq1 and seq2 using global alignment and print out the alignments and their scores
# 2. Store the BEST alignment in variable: alignment

seq1 = "AAAGGTCCTAGTTGG"
seq2 = "AAAGTCCTAGTTGG"

aligner = Align.PairwiseAligner()
alignments = aligner.align(seq1, seq2)

for aln in alignments:
    print(aln)
    print(f"Score: {aln.score}")

alignment = alignments[0]
```

## Part 3: Substitution Matrix

Load the **BLOSUM62** matrix and use it in the `aligner_matrix`.

Store the best alignment in:

    matrix_alignment

```python
# TODO: use BLOSUM62, re-align the seq1 and seq2, store the best alignment score

matrix = substitution_matrices.load("BLOSUM62")
aligner_matrix = Align.PairwiseAligner()
aligner_matrix.substitution_matrix = matrix

alignments_matrix = aligner_matrix.align(seq1, seq2)
matrix_alignment = alignments_matrix[0]
score = matrix_alignment.score
score
```

## Part 5: Alignment Output Regions of Similarities

Use the following sequences:

- target = "GAACTTT"
- query = "GATTT"

```python
# TODO: 
# 1. align target and query sequences, store their alignments in the alignments variable, print out the each alignment and its score.

target = "GAACTTT"
query = "GACTT"

aligner = Align.PairwiseAligner()
alignments = aligner.align(target, query)

# Print each alignment and its score
for alignment in alignments:
    print(alignment)
    print(f"Score: {alignment.score}\n")
```

```python
### BEGIN TESTS

assert isinstance(alignments, Align.PairwiseAlignments), f"Expected alignments to be a Biopython PairwiseAlignments object, got {type(alignments)}"
assert len(alignments) == 6, "Alignments should contain 6 different alignments"

### END TESTS
print("Test passed!")
```

```python
# TODO: store the coordinates or the range(numpy array) of regions of similarities between the target and query sequences for each alignment in alignments in a list variable x.

x = []
for alignment in alignments:
    coords = alignment.coordinates
    # Remove the last coordinate if odd number of values to make pairs of (start, end)
    if coords.shape[1] % 2 == 1:
        coords = coords[:, :-1]
    # Reshape to (n_sequences, n_regions, 2) where each region is (start, end)
    coords_reshaped = coords.reshape(coords.shape[0], -1, 2)
    x.append(coords_reshaped)
```

```python
### BEGIN TESTS

assert isinstance(x, list), "x should be a list"
assert len(x) != 0, "x should not be empty"
assert [i.shape for i in x] == [(2, 2, 2), (2, 2, 2), (2, 3, 2), (2, 3, 2), (2, 3, 2), (2, 3, 2)], "The shapes of the aligned regions should match expected values"
### END TESTS
print("Test passed!")
```
